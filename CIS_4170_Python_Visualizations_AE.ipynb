Salary v.s. Median Worker Pay by Industry


```python
df = pd.read_csv('ceo_data_sp500.csv')
```

```python
cols = ['industry', 'salary', 'median_worker_pay']
df = df[cols]
```

```python
df = df.loc[df['salary'].notnull() & df['median_worker_pay'].notnull()]
```

```python
df['salary'] = df['salary'].apply(lambda x: x.replace('$', '').replace(',', '').strip()).astype(float)
df['median_worker_pay'] = df['median_worker_pay'].apply(lambda x: x.replace('$', '').replace(',', '').strip()).astype(float)
```

```python
grouped = df.groupby('industry').median().reset_index()
```

```python
grouped = grouped.sort_values('salary', ascending=True)
```

```python
fig, ax = plt.subplots(figsize=(10, 15))
ax.barh(grouped['industry'], grouped['salary'], label='CEO Salary', color='orange')
ax.barh(grouped['industry'], grouped['median_worker_pay'], label='Median Worker Pay', color='blue')
ax.set_xlabel('Salary ($)')
ax.set_ylabel('Industry')
ax.set_title('Comparison of Median CEO Salary and Median Worker Pay by Industry')
ax.legend()
plt.show()
```

```python
# Group the dataframe by industry and compute the median CEO salary and median worker pay
industry_data = df.groupby('industry').median()

# Create a scatter plot to compare CEO salary and median worker pay by industry
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(industry_data['salary'], industry_data['median_worker_pay'], s=100, alpha=0.8)

# Add labels and title to the plot
ax.set_xlabel('CEO Salary ($)', fontsize=12)
ax.set_ylabel('Median Worker Pay ($)', fontsize=12)
ax.set_title('Comparison of CEO Salary and Median Worker Pay by Industry', fontsize=16)

# Show the plot
plt.show()
```

```python
# Group the dataframe by industry and calculate the median salary and median worker pay for each industry
industry_stats = df.groupby('industry').median()

# Create a scatterplot to compare CEO Pay and Median Worker Pay by Industry
fig, ax = plt.subplots(figsize=(10,8))
colors = plt.cm.tab20.colors
for i, ind in enumerate(industry_stats.index):
    x = industry_stats.loc[ind, 'salary']
    y = industry_stats.loc[ind, 'median_worker_pay']
    ax.scatter(x, y, s=100, label=ind, color=colors[i])
    
ax.set_xlabel('CEO Salary ($)', fontsize=12)
ax.set_ylabel('Median Worker Pay ($)', fontsize=12)
ax.set_title('Comparison of CEO Pay and Median Worker Pay by Industry', fontsize=16)
ax.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a pandas dataframe
df = pd.read_csv('ceo_data_sp500.csv')

# Select the columns we're interested in
cols = ['industry', 'salary', 'median_worker_pay']

# Filter out any rows where the salary is null
df = df.loc[df['salary'].notnull(), cols]

# Convert the salary and median worker pay columns to millions
#df['salary'] = df['salary'].replace('$','').replace(',', '').astype('float')/ 1000000
#df['median_worker_pay'] = df['median_worker_pay'].replace('$','').replace(',', '').astype('float')// 1000000
df['salary'] = df['salary'].apply(lambda x: float(x.strip('$').replace(',', '')) / 1000000)
df['median_worker_pay'] = df['median_worker_pay'].apply(lambda x: float(x.strip('$').replace(',', '')) / 1000000)

# Group the data by industry and calculate the mean of salary and median worker pay
industry_data = df.groupby('industry').mean()

# Create a scatter plot to compare CEO Pay and Median Worker Pay by Industry
fig, ax = plt.subplots(figsize=(10, 8))
colors = plt.cm.tab20.colors
for i, ind in enumerate(industry_stats.index):
    x = industry_stats.loc[ind, 'salary']
    y = industry_stats.loc[ind, 'median_worker_pay']
    ax.scatter(x, y, s=100, label=ind, color=colors[i])
ax.scatter(industry_data['salary'], industry_data['median_worker_pay'], s=industry_data['salary']**2, alpha=0.5)
ax.set_xlabel('CEO Salary ($ Millions)', fontsize=12)
ax.set_ylabel('Median Worker Pay', fontsize=12)
ax.set_title('Comparison of CEO Salary and Median Worker Pay by Industry', fontsize=16)

ax.legend()

# Show the plot
plt.show()
```
